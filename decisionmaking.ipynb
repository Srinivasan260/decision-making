```python
import pandas as pd

data = pd.read_csv('decision.csv')

data.head()
```

```python
inputs = data.drop('salary greater than 100k',axis='columns')
target=data['salary greater than 100k']
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
inputs['company_n']=le_company.fit_transform(inputs['company'])
inputs['job_n']=le_job.fit_transform(inputs['job'])
inputs['degree_n']=le_degree.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
datas=inputs.drop(['company','job','degree'],axis='columns')
```

```python
datas
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(datas[['company_n','job_n','degree_n']],target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
X_test
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```
